```python
# Combine KS, EZCAP, HAX claims
path_file_ks = '../data/csv/HEDIS KS Claims.csv'
path_file_ezcap = '../data/csv/HEDIS EZCAP Prof hp_claims.csv'
path_file_hax = '../data/csv/HEDIS HAX HP_Claims.csv'

path_export = '../output/HP_Claim_20211108.txt'

dtype = {'provnbr':str, 'rev':str, 'units':str, 'provspec':str, 'billtype':str, 'pos':str}

df_ks = pd.read_csv(path_file_ks, low_memory=False, dtype=dtype)
df_ezcap = pd.read_csv(path_file_ezcap, low_memory=False, dtype=dtype)
df_hax = pd.read_csv(path_file_hax, low_memory=False, dtype=dtype)

df = pd.concat([df_ks, df_ezcap, df_hax])
df['fromdt'] = df['fromdt'].str.strip()
df['thrudt'] = df['thrudt'].str.strip()

df.to_csv(path_export, sep='|', index=False)

df.head(3)
```
