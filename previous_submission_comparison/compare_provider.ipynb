# Provider Compare Current Pull with Last Submission

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
file_name = 'HP_Provider'

path_prev = f'../data/past_submission/{file_name}_20210921.txt'
path_now = f'../output/{file_name}_20211025.txt'

dtype = {'provfax1':str,
         'provspec1':str,
         'provspec2':str,
         'provspec3':str,
         'provspec4':str,
         'provspec5':str,
         'provspec6':str,
         'provspec7':str,
         'provspec8':str,
         'provspec9':str,
         'provspec10':str}

df_prev = pd.read_csv(path_prev, sep='|', low_memory=False, dtype=dtype)
df = pd.read_csv(path_now, sep='|', low_memory=False, dtype=dtype)
```

```python
print(df_prev.shape)
display(df_prev.head(3))
```

```python
print(df.shape)
display(df.head(3))
```

```python
df_inspect = df[['npi']].drop_duplicates()\
    .merge(df_prev[['npi']].drop_duplicates(), on='npi', how='left', indicator=True)

n = df_inspect.query('_merge == "left_only"')['npi'].count()
print(f'Current submission not found in previous submission: {n}')
```

```python
df_inspect['_merge'].value_counts()
```
