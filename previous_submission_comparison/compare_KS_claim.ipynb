# KS Claim Compare Current Pull with Last Submission

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
file_name = 'HP_Claim_KS'

path_prev = f'../data/past_submission/{file_name}_20210921.txt'
path_now = f'../output/{file_name}_20211025.txt'

dtype = {'provnbr':str}

df_prev = pd.read_csv(path_prev, sep='|', low_memory=False, dtype=dtype)
df = pd.read_csv(path_now, sep='|', low_memory=False, dtype=dtype)
```

```python
print(df_prev.shape)
display(df_prev.head(3))
```

```python
print(df.shape)
display(df.head(3))
```

```python
df_inspect = df[['claimid', 'fromdt']].drop_duplicates()\
    .merge(df_prev[['claimid']].drop_duplicates(), on='claimid', how='left', indicator=True)

df_inspect['year'] = df_inspect['fromdt'].str[-4:].astype(int)
df_inspect['month'] = df_inspect['fromdt'].str[:2].astype(int)

n = df_inspect.query('_merge == "left_only"')['claimid'].count()
print(f'Current submission not found in previous submission: {n}')
display(df_inspect.query('_merge == "left_only"').groupby(['year'])['claimid'].count().to_frame())
display(df_inspect.query('_merge == "left_only"').groupby(['year', 'month'])['claimid'].count().to_frame())
```

```python
df_inspect['_merge'].value_counts()
```

```python
df_inspect = df_prev[['claimid', 'fromdt']].drop_duplicates()\
    .merge(df[['claimid']].drop_duplicates(), on='claimid', how='left', indicator=True)
df_inspect['year'] = df_inspect['fromdt'].str[-4:].astype(int)
df_inspect['month'] = df_inspect['fromdt'].str[:2].astype(int)

n = df_inspect.query('_merge == "left_only"')['claimid'].count()
print(f'Previous submission not found in current submission: {n}')
display(df_inspect.query('_merge == "left_only"').groupby(['year'])['claimid'].count().to_frame())
display(df_inspect.query('_merge == "left_only"').groupby(['year', 'month'])['claimid'].count().to_frame())
```
